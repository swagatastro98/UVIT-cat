```python
import numpy as np
import pandas as pd
from glob import glob
from astropy.io import fits
from astropy.table import Table
```

```python
def merge_all(path):
    
    files = glob(f'{path}*.fits')
    
    dflist = []
    
    for x in files:
    
        img = fits.open(x)
    
        img.info()

        data = img[1].data
        
        newdf = Table(data)
        df = newdf.to_pandas()
        
        dflist.append(df)
        
        for m in df.columns:
            
            print('For four dataframes: ')
            print(f'{m}:',df[m].isna().sum())
        
    final_df = pd.concat(dflist,ignore_index=True)
    
    print(dflist[1].columns)
#     print(final_df['F1_F_A'].isna().sum())
   
#     for y in final_df.columns:
        
#         print('For final dataframe')
#         print(final_df[y].isna().sum())
    
    
merge_all('/Users/swagat98/MAIN_UVIT_cat/')
```
